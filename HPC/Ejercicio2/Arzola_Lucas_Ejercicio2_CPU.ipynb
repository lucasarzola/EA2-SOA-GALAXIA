#3 Desarrollo

```
# ------------------------------------------------------------
#@title 3.1 Parámetros de ejecución { vertical-output: true }

tam_mat =   -2#@param {type: "number"}

# ------------------------------------------------------------
from datetime import datetime
tiempo_total = datetime.now()
import numpy
import copy
# --------------------------------------------
# Definición de función que transforma el tiempo en  milisegundos 
tiempo_en_ms = lambda dt:(dt.days * 24 * 60 * 60 + dt.seconds) * 1000 + dt.microseconds / 1000.0
# --------------------------------------------

# CPU - Defino la memoria de la matriz en cpu.
try :
  matriz_cpu = numpy.random.randn(tam_mat,tam_mat) *10
  matriz_cpu = matriz_cpu.astype(numpy.int32())
#Realizo la transpuesta 

  tiempo_bucle = datetime.now()
  matriz_r = numpy.random.randn(tam_mat,tam_mat) 
  matriz_r = matriz_r.astype(numpy.int32())
  for i in range(len(matriz_cpu[0])) :
    for j in range(len(matriz_cpu)):
      matriz_r[j][i]=matriz_cpu[i][j]
  tiempo_bucle = datetime.now() - tiempo_bucle

#informo Resultados.

  print("Matriz 1: ")
  print(matriz_cpu)
  print("Resultado de la Transposición de la matriz")
  print(matriz_r)

  tiempo_total = datetime.now() - tiempo_total

  print("Tiempo Total: ", tiempo_en_ms( tiempo_total ), "[ms]" )
  print("Tiempo bucle: ", tiempo_en_ms( tiempo_bucle ), "[ms]" )


    
except:
  print("No se puede poner tamaños de matriz negativos")
  exit()
```
